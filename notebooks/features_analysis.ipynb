```python
import os
import pandas as pd
```

```python
data_dir = "../data/raw/physionet.org/files/ptb-xl/1.0.1"
data_path = os.path.join(data_dir, "ptbxl_database.csv")
```

```python

table = pd.read_csv(data_path, index_col="ecg_id")
technical_features = ["patient_id", "nurse", "device", "recording_date", "report", "scp_codes", "strat_fold", "filename_lr", "filename_hr"]
table = table.drop(columns=technical_features)
table.head()
```

```python
print("Nan count by feature")
table.isna().sum()
```

```python
unique_values = {col: (len(table[col].unique()), table[col].dtype) for col in table.columns}
print("UNIQUE VALUES:\n")

for name, (count, dtype) in unique_values.items():
    print(f"{name}: {count} -- {dtype}")
```
